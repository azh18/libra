## 获取所有指数的meta信息

包括：
- 发布日期
- 名称

```python
class IndexMeta:
    def __init__(self, code, meta_pd_series):
        self.code = code
        self.launch_date = str(meta_pd_series["LAUNCHDATE"].date())
        self.name = meta_pd_series["SEC_NAME"]
    def __str__(self):
        return "%s: [%s, %s]" % (self.code, self.launch_date, self.name)
        
```

```python
# code_list is List[Str]
# return Dict[Code]SecurityMeta, e.g. {"399989.SZ": SecurityMeta{"2014-10-31", "中证医疗"}}
def get_index_meta(code_list):
    ret_code, df = w.wss(code_list, "launchdate,sec_name", usedf=True)
    if ret_code != 0:
        raise Exception("ret code=", ret_code)
    data = {}
    print(df)
    for code in code_list:
        data[code] = IndexMeta(code, df.loc[code])
    return data
```

```python
etf_codes = [
    '159828.SZ', "515220.SH", "159939.SZ", "159920.SZ", "512980.SH", "512880.SH", "512480.SH", "512660.SH", "515050.SH",
    "515700.SH", "512800.SH", "510050.SH", "510300.SH", "510500.SH", "588000.SH"
]


# Return: Dict[index_code] -> etf_code
def etf_to_index(etf_codes):
    index_etf_map = {}
    err_code, d = w.wss(etf_codes, "fund_trackindexcode", usedf=True)
    if err_code != 0:
        raise Exception("etf to index error: %d" % err_code)
    for etf_code, index_code in d["FUND_TRACKINDEXCODE"].iteritems():
        index_etf_map[index_code.upper()] = etf_code
    return index_etf_map

index_etf_map = etf_to_index(etf_codes)
index_etf_map.keys()
```

```python
class IndexMeta:
    def __init__(self, code, etf_code, meta_pd_series):
        self.code = code
        self.etf_code = etf_code
        self.launch_date = str(meta_pd_series["LAUNCHDATE"].date())
        self.name = meta_pd_series["SEC_NAME"]
        
    def __str__(self):
        return "%s: [%s, %s]" % (self.code, self.launch_date, self.name)
    
    # ('code', '中文名', etf_code')
    def to_tuple(self):
        return (self.code, self.name, self.etf_code)
    
# code_list is List[Str]
# return Dict[Code]SecurityMeta, e.g. {"399989.SZ": SecurityMeta{"2014-10-31", "中证医疗"}}
def get_index_meta(code_list, index_etf_map):
    ret_code, df = w.wss(code_list, "launchdate,sec_name", usedf=True)
    if ret_code != 0:
        raise Exception("ret code=", ret_code)
    data = {}
    print(df)
    for code in code_list:
        data[code] = IndexMeta(code, index_etf_map[code], df.loc[code])
    return data

def get_latest_series_data(code, start_time, end_time):
    ret_code, d = w.wsd(code, "pe,pe_ttm,pb_lf,val_pb_percentile,val_pe_percentile,ps,roe_ttm,open,close,high,low,pct_chg,volume", start_time, end_time, "ruleType=10;PriceAdj=F", usedf=True)
    if ret_code != 0:
        raise Exception("get_latest_series_data for %s error: %d" % (code, ret_code))
    return d

def get_today_date():
    import time
    return time.strftime("%Y-%m-%d", time.localtime()) 

# Return: Dict[index_code] -> etf_code
def etf_to_index(etf_codes):
    index_etf_map = {}
    err_code, d = w.wss(etf_codes, "fund_trackindexcode", usedf=True)
    if err_code != 0:
        raise Exception("etf to index error: %d" % err_code)
    for etf_code, index_code in d["FUND_TRACKINDEXCODE"].iteritems():
        index_etf_map[index_code.upper()] = etf_code
    return index_etf_map
```

```python
import os
import pickle
import pandas as pd

data_dir = os.path.join(os.path.curdir, "data")
try:
    os.makedirs(data_dir)
except:
    pass

def get_csv_path(code):
    return os.path.join(data_dir, "%s.csv" % code)

def read_pd_from_csv(code):
    return pd.read_csv(get_csv_path(code), index_col="date")

def dump_pd_to_csv(code, df):
    df.to_csv(get_csv_path(code), index_label="date")
    
def dump_index_meta_map(index_meta_map):
    with open(os.path.join(data_dir, "index_meta.pkl"), "wb") as f:
        pickle.dump(index_meta_map, f)

def load_index_meta_map():
    with open(os.path.join(data_dir, "index_meta.pkl"), "rb") as f:
        index_meta_map = pickle.load(f)
    return index_meta_map
```
